```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("correct_twitter_201904.tsv", sep='\t', header=0)
```

```python
df.describe()
```

```python
df['ts1'] = pd.to_datetime(df['ts1'])
```

```python
# 1. How many tweets were posted containing the term on each day?
```

```python
def query_tweets_by_term(term):
    # Filter tweets that contain the term in the text (assuming 'text' column contains the tweet content)
    filtered_df = df[df['text'].str.contains(term, case=False, na=False)]
```

```python
tweets_per_day = df.groupby(df['ts1'].dt.date).size()
print("Tweets per day containing the term '{}':\n".format('music'), tweets_per_day)
```

```python
# 2. Number of unique users who posted tweets containing the term
```

```python
unique_users = df['author_handle'].nunique()
print(unique_users)
```

```python
# 3. Average number of likes for tweets containing the term
```

```python
avg_likes = df['like_count'].mean()
print(avg_likes)
```

```python
# 4. Place IDs where tweets originated from
```

```python
place_ids = df['place_id'].dropna().unique()
print(place_ids)
```

```python
# 5. Distribution of tweet posting times (hours of the day)
```

```python
times_of_day = df['ts1'].dt.hour.value_counts().sort_index()
print (times_of_day)
```

```python
 # 6. User who posted the most tweets containing the term
```

```python
top_user = df['author_handle'].value_counts().idxmax()
print (top_user)
```

```python
# Combine the results into a dictionary
```

```python
result={
        "tweets_per_day": tweets_per_day,
        "unique_users": unique_users,
        "avg_likes": avg_likes,
        "place_ids": place_ids,
        "times_of_day": times_of_day,
        "top_user": top_user

        }
result

```
